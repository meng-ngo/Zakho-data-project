```python
import os
```

```python
with open(os.path.join('data', 'uni-content.html'), 'r') as f:
    text = f.read()
    print(text)
```

```python
from bs4 import BeautifulSoup
```

```python
# Remove HTML tag using BeautifulSoup

soup = BeautifulSoup(text, 'html5lib')
print(soup.get_text())
```

```python
paragraphs = soup.findAll('p')
```

```python
info = dict()
```

```python
for para in paragraphs:
    sibling = para.findPreviousSibling('h2')
    if sibling:
        title = sibling.get_text().strip()
        if title not in info: info[title] = list()
        info[title].append(para.get_text().strip())
        print(sibling.get_text().strip())
        print(para.get_text().strip())
```

```python
info
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentences = sent_tokenize(info['Zakho City'][0])
```

```python
sentences
```

```python
articles_to_sentences = dict()
```

```python
for para in info:
    articles_to_sentences[para] = []
    for article in info[para]:
        articles_to_sentences[para].append(sent_tokenize(article))
```
